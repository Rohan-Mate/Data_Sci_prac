# Pandas export

to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
# use reset_index -> convert pandas series into dataframe
```

```python
temp_df
```

```python
temp_df.to_csv('batsman_runs.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

3. to_html

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df
```

```python
df.query('batsman_runs == 6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count')
```

```python
df.query('batsman_runs == 6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count').to_html('sixes_heatmap.html')
```

4. to_json

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
